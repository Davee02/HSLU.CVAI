# Faltung 1D

## Initialisierung

```python
import numpy as np
import cv2 as cv
from matplotlib.pyplot import *
```

## Folgen (als Funktionen) definieren

### Bereich definieren

```python
Rmin = -5
Rmax = 15
R = np.array (range (Rmin, Rmax + 1))
print (R)
```

### Erste Folge

```python
def X (k):
    if ((0 <= k) & (k < 4)):
        return 1
    return 0
```

```python
x = np.array ([X (i) for i in R])
print (x)
stem (R, x)
```

### Zweite Folge

```python
def Y (k):
    if ((k == 0) | (k == 2)):
        return 1
    if (k == 1):
        return 2
    return 0
```

```python
y = [Y (i) for i in R]
print (y)
stem (R, y)
```

## Konvolution

### Ein Element berechnen

```python
def Conv0 (X, Y, k):
    s = 0
    for i in range (Rmin, Rmax + 1):
        s = s + X (i)* Y (k - i)
    return s
```

```python
print (Conv0 (X, Y, 0))
print (Conv0 (X, Y, 1))
print (Conv0 (X, Y, 2))
```

### Alle Elemente berechnen

```python
def Conv (X, Y):
    n = R.shape [0]
    xy = np.zeros (n, dtype = int)
    for i in R:
        k = R [i]
        xy [i] = Conv0 (X, Y, k)
    return xy
```

```python
Z = Conv (X, Y)
print (Z)
stem (R, Z)
```

### Konvolution mit Python-Funktion berechnen

```python
x = np.array ([1, 2, 3, 4, 3, 2, 1])
y = np.array ([1, 2, 1])
z = np.convolve (x, y)

print (x)
print (y)
print (z)

subplot (2, 2, 1); stem (x)
subplot (2, 2, 2); stem (y)
subplot (2, 2, 3); stem (z)
```

## Weiteres Beispiel

```python
x = np.array ([1, 3, 5, 7, 6, 4, 2])
y = np.array ([0, 1, 0])
z = np.convolve (x, y)

print (x)
print (y)
print (z)

subplot (2, 2, 1); stem (x)
subplot (2, 2, 2); stem (y)
subplot (2, 2, 3); stem (z)
```
